```python
# 不同结构的数据框，按照一定的条件进行合并
# merge(x, y, left_on, right_on)
# 分别为第一个数据框， 第二个数据框， 第一个数据框匹配的列， 第二个数据框匹配的列
# 返回值 DataFrame
```

```python
import pandas;
from pandas import read_csv

items = read_csv(
    "./data1.csv", 
    sep='|', 
    names=['id', 'comments', 'title']
)
items
```

```python
prices = read_csv(
    "./data2.csv", 
    sep='|', 
    names=['id', 'oldPrice', 'nowPrice']
)
prices
```

```python
itemPrices = pandas.merge(
    items, 
    prices, 
    left_on='id', 
    right_on='id'
)
itemPrices
```
