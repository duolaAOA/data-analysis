```python
import matplotlib
from pandas import read_csv
from matplotlib import pyplot as plt

font = {
    'family' : 'SimHei'
}
matplotlib.rc('font', **font)

data = read_csv('.\data.csv')
data
```

```python
plt.hist(data['购买用户数'])
plt.show()
```

```python
plt.hist(data['购买用户数'], bins=20)
plt.show()
```

```python
plt.hist(data['购买用户数'], bins=20, cumulative=True)
plt.show()
```
