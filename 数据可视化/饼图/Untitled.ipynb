```python
import numpy
import matplotlib
from pandas import read_csv
import matplotlib.pyplot as plt
data = read_csv('.\data.csv')
data
```

```python
# groupby 分组统计
gb = data.groupby(
    by=['通信品牌'], 
    as_index=False
)['号码'].agg({
    '用户数':numpy.size
})
gb
```

```python
font = {
    'family' : 'SimHei'
}

matplotlib.rc('font', **font)
plt.pie(gb['用户数'], labels=gb['通信品牌'], autopct='%.2f%%')
plt.show()
```
