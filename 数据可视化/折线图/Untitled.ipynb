```python
import pandas
from pandas import read_csv
from matplotlib import pyplot as plt
data = read_csv('.\data.csv')
data
```

```python
#对日期格式进行转换
data['购买日期']=pandas.to_datetime(data['日期'])
```

```python
#'-'	顺滑的曲线
plt.plot(data['购买日期'], data['购买用户数'], '-', color='r')
plt.show()
```
